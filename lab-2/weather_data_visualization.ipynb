```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/london_weather.csv')
df.info()
```

```python
df.head()
```

```python
# Question: Visualize the change in average temperature over the years in London

# Answer:
# Step1: Find year from date and store it in year column
dates = df['date']
years = []
for date in dates:
    year = int(str(date)[:4])
    years.append(year)
df['years'] = years

# Step2: Calculate the average temperature of each year from 1979 to 2020
average_temps = df.groupby('years')['mean_temp'].mean()

print(type(average_temps))

# Step3: Draw line plot
plt.title('Year wise Mean Temperature of London (1979-2020)')
plt.ylabel('Mean Temperature')
plt.xlabel('Years')
plt.plot(average_temps)
```
