```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/london_weather.csv')
df.info()
```

```python
df.head()
```

```python
# Question: Visualize the change in average temperature over the years in London

# Answer:
# Step1: Find year from date and store it in year column
dates = df['date']
years = []
for date in dates:
    year = int(str(date)[:4])
    years.append(year)
df['years'] = years

# Step2: Calculate the average temperature of each year from 1979 to 2020
average_temps = df.groupby('years')['mean_temp'].mean()

print(type(average_temps))

# Step3: Draw line plot
plt.title('Year wise Mean Temperature of London (1979-2020)')
plt.ylabel('Mean Temperature')
plt.xlabel('Years')
plt.plot(average_temps)
```

```python
# Question: Visualize the relationship between temperature and pressure

# Answer:

# Step1: Draw scatter plot of temperature vs pressure
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.scatter(df['mean_temp'], df['pressure'])

# Insight: There is no significant relationship between mean temperature and pressure
```

```python
# Question: Visualize distribution of precipitation over the year in London

# Answer:

# Step1: Find month from date and store it in month column
dates = df['date']
months = []
for date in dates:
    month = int(str(date)[4:6])
    months.append(month)
df['months'] = months

# Step2: Visualize the data using histogram
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.violinplot(x='months', y='precipitation', data=df)
plt.title('Monthly Distribution of Precipitation in London')
plt.xlabel('Month')
plt.ylabel('Precipitation (mm)')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```
