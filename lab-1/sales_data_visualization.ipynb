```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/Auto Sales data.csv')
```

```python
df.head()
```

```python
# Visualize the no. of order month wise
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
no_of_orders = [0] * 12
orderdates = df['ORDERDATE']
for date in orderdates:
    month = int(date.split('/')[1])
    no_of_orders[month - 1] += 1

plt.plot(months, no_of_orders)
plt.title('Sales Monthwise')
plt.xlabel('Months')
plt.ylabel('No. of orders')
plt.show()
```
