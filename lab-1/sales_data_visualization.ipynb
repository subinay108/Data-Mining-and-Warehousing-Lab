```python
# Drop some random values to reduce the dataset by 70%
from random import random
def reduce_dataset(df, frac):
    delete_index = []
    for i in range(len(df)):
        if random() > frac:
            delete_index.append(i)
    return df.drop(delete_index)

new_df = reduce_dataset(df, 0.1)
```
